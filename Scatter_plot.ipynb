```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("DataPreparation/Pokemon_mod_v1.csv")
cols = list(data.columns)
Types = data['Type'].unique()
Types.sort()
colors = ["#a6b91a","#705746","#6f35fc","#f7d02c","#d685ad","#c22e28","#ee8130","#a98ff3","#735797","#7ac74c","#e2bf65","#96d9d6","#a8a77a","#a33ea1","#f95587","#b6a136","#b7b7c3","#6390f0"]
data = data[cols[:2]+cols[3:5]]
# grouped_data = np.array([data.loc[data["Type"]=="Bug"].drop(columns=['Type'])])
# a = np.array([2,2])
# np.vstack([grouped_data,a])

plt.figure(figsize=(8, 8))
typy = ["Bug","Dragon","Fire","Rock"]
plt.xlabel('Total Defense',fontsize="16")
plt.ylabel('Total Attack',fontsize="16")
for typ,color in zip(Types,colors):
    if typ not in typy:continue
#     typ_data = data.loc[(data["Type"]==typ)&(~data["Name"].str.contains("Mega"))&((data["Total Attack"] >= 400)|(data["Total Defense"] >= 400))].drop(columns=['Type'])
    typ_data = data.loc[data["Type"]==typ].drop(columns=['Type'])
    name = typ_data['Name']
    attack = typ_data['Total Attack']
    defense= typ_data['Total Defense']
    plt.scatter(defense,attack,50,alpha=0.8,c=color,label=typ)
#     for n,d,a in zip(name,defense,attack):
#         plt.annotate(n,(d,a))
    
plt.legend(bbox_to_anchor=(1,1))

plt.savefig('Scatter_plot_for_a_few_types_of_pokemon.pdf')
plt.show()
```
