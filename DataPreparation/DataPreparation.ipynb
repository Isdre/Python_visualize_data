```python
import pandas as panda
dp = panda.read_csv("Pokemon.csv")
dp.head(5)
```

```python
dp.info()
```

```python
dp['Total Attack'] = dp.iloc[:,[6,8,10]].sum(axis=1)
dp['Total Defense'] = dp.iloc[:,[5,7,9]].sum(axis=1)
dp.head(5)
```

```python
dp.info()
```

```python
cols = list(dp.columns)
print(cols)
# dp = dp[cols[0:5] + cols[13:15] + cols[5:13]]
# dp.head(5)
```

```python
dpm = dp[cols[0:5] + cols[13:15] + cols[5:13]]
dpm.rename(columns={"#":"ID"},inplace=True)
dpm.head(5)
```

```python
dpm.to_csv("Pokemon_mod.csv",index=False)
```

```python
#dpm.loc[dpm['Name'].str.contains("Mega")]
types = dpm['Type 1']
types = types.unique()
with open('types.txt',"w") as types_file:
    for typ in types:
        types_file.write(f'{typ}\n')
# dpm.loc[dpm['Type_1'] == types[0]]
```

```python
dpm1 = dpm.drop(columns=['Type 2'])
dpm2 = dpm.drop(columns=['Type 1'])
dpm1.rename(columns={"Type 1":"Type"},inplace=True)
dpm2.rename(columns={"Type 2":"Type"},inplace=True)
dpm2 = dpm2.dropna()
dpm2.head(5)
```

```python
dpm_f = panda.concat([dpm1,dpm2])
dpm_f.drop(columns=['ID'],inplace=True)
dpm_f.head(5)
dpm_f.to_csv("Pokemon_mod_v1.csv",index=False)
```

```python
for typ in types:
    a = dpm_f.loc[dpm_f['Type'] == typ]
    a.reset_index(drop=True,inplace=True)
    a.to_csv("{}.csv".format(typ))
# a = dpm_f.loc[dpm_f['Type'] == types[0]]
# a.reset_index(drop=True,inplace=True)
# print(a.head(5))
# print("{}.csv".format(types[0]))
```
